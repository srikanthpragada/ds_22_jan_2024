```python
def isprime(n):
    for i in range(2,n//2 + 1):
        if n % i == 0 :
               return False
            
    return True
```

```python
# Returns a list of booleans 
def prime(a):
    lst = []
    for n in a:
         lst.append(isprime(n))   # Append true or false 
        
    return lst  # List of booleans 
```

```python
a = np.random.randint(1,100,10)
a
```

```python
print(prime(a))
```

```python
a[prime(a)]
```

```python
a[[isprime(n) for n in a]]  # List comprehension for boolean indexing 
```
